### Graph plotting w.r.t. time in hours and the vehicle type

```python
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M:%S')
data['hour'] = data['Time'].dt.hour

time_vehicle = data.groupby(['hour', 'Vehicle Type']).size().unstack(fill_value=0)
```

```python
time_vehicle.plot(kind='bar', stacked=True, figsize=(11,6))
plt.title("Vehicle Type Bookings by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=360)
plt.show()
```
